## Importing necessary libraries

```python
import requests
import pandas as pd
import time
import random
import re
import numpy as np
import _pickle as pickle
from tqdm import tqdm_notebook as tqdm
from bs4 import BeautifulSoup as bs
```

# Gathering the user bio data
### BeautifulSoup is used to navigation
### BeautifulSoup is a python library for pulling data out of HTML or XML files

```python
# Randomizing the refresh rate
seq = [i/10 for i in range(8,20)] #Represent the number of seconds to wait for refreshing the page [0.8-2.0 secs]

# Creating a list of bios
biolist = []

# Gathering bios by looping and refreshing the web page 
for _ in tqdm(range(1000)):
    
    # Refreshing the page
    page = requests.get("https://www.fakepersongenerator.com/user-biography-generator")
    soup = bs(page.content)
    
    try:
        # Getting the bios
        bios = soup.find('div', class_='row no-margin for-sign').find_all('p')

        # Adding to a list of the bios
        biolist.extend([re.findall('"([^"]*)"', i.text) for i in bios])
    except:
        pass
    
    # Sleeping 
    time.sleep(random.choice(seq))
```

```python
# Creating a DF from the bio list
bio_df = pd.DataFrame(biolist, columns=['Bios'])
bio_df
```

# Adding other attributes to the profiles

```python
# List of Dating questions according to the questionnaire
attr = ['Drinks',
      'Perfume',
      'Vacation',
      'Personal Style',
      'Music',
      'Hobby',
      'Food',
      'Scent',
      'Feel']
```

```python
# Creating a DF of the categories
attr_df = pd.DataFrame(columns=attr)
```

```python
#Randomizing the attributes to numerically represent the categorical data
for i in attr_df.columns:
    
    # Range of numbers to represent different categories of each attributes
    attr_df[i] = np.random.randint(0,13, bio_df.shape[0])
    
    # The numbers represent a specific choice within the categories
    # For e.g., your preffered genre under the Music, your favourite hobby, etc.
```

```python
# Viewing the randomized data
attr_df
```

```python
# Joining the two dataframes ro create the final dataframe
final_df = bio_df.join(attr_df)
final_df
```

# Exporting the Final Random User DataFrame

```python
with open("profiles.pkl", "wb") as fp:
    pickle.dump(final_df, fp)
```
